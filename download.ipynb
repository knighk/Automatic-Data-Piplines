```python
import csv
import requests

with open('2014_4.csv', newline='', encoding="utf-8") as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for line in reader:
        saveas = "/home/kke1/data/sec/2014/QTR4/"+'-'.join([line[0], line[1].replace('/', '-'),  line[3].replace('/','_')])+".html"
        # saveas = '-'.join([line[0],line[4].replace('/','_'),"10Q"])
        # saveas= "J:\\data\\edgar\\2014.3\\"+ saveas+".html"

        # Reorganize to rename the output filename.
        url = 'https://www.sec.gov/Archives/' + line[5].strip()
        with open(saveas, 'wb') as f:
            f.write(requests.get('%s' % url).content)
            print(url, 'downloaded and wrote to text file')
            
print("done!")
```
